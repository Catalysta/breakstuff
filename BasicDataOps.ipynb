## Train test splits and correlation analysis

```python
X, y = dat.drop(['id', 'diagnosis'], axis = 1), dat[['diagnosis']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=42)
X_test, X_holdout, y_test, y_holdout = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

```python
print(X_train.shape,
      X_test.shape,
      X_holdout.shape,
      y_train.shape,
      y_test.shape,
      y_holdout.shape)
```

```python
def hist_comp(X_train, y_train, colname):
    
    plt.hist(X_train.loc[y_train['diagnosis'] == "M", colname], label='malignant')
    plt.hist(X_train.loc[y_train['diagnosis'] == "B",colname], label='benign')
    plt.legend(loc='upper right')
    plt.title(colname)
    plt.show()

for cols in X_train.columns:
    hist_comp(X_train, y_train, cols)
```

```python
cols_to_drop = ['texture_mean'
               , 'smoothness_mean'
               , 'symmetry_mean'
               , 'fractal_dimension_mean'
               , 'texture_se'
               , 'smoothness_se'
               , 'concavity_se'
               , 'concave points_se'
               , 'fractal_dimension_se']

X_train = X_train.drop(cols_to_drop, axis = 1)
```

```python
test = mannwhitneyu(X_train.loc[y_train['diagnosis'] == "M", 'concave points_mean']
                   , X_train.loc[y_train['diagnosis'] == "B", 'concave points_mean'])
test[1]
```

```python
def test_meanshift(X_train, y_train, colname):
    #check for mean shift using wilcoxan rand sum test
    test = mannwhitneyu(X_train.loc[y_train['diagnosis'] == "M", colname]
                        , X_train.loc[y_train['diagnosis'] == "B", colname])
    return(test[1])

pvals = list(map(lambda x: test_meanshift(X_train, y_train, x), X_train.columns))
result = dict(zip(X_train.columns, pvals))
result
```
